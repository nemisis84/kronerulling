```python
import matplotlib.pyplot as plt
```

```python
# Set up the data for the bar chart
linjeforeningerDict = {
    'Abakus': 10000,
    'Thilde':2365,
    'Hybrida':9212,
    'Janus':5678,
    'Vivas':15432,
    'Delta': 2563
  }
plt.figure(figsize=(8, 5))
# Sort the data by the values on the x-axis
data = [(value, label) for label, value in linjeforeningerDict.items()]
data.sort()

# Extract the sorted values on the x-axis and the corresponding labels on the y-axis
x, y = zip(*data[len(data)-5:]) #Slice for the 5 largest contributors

# Create the bar chart
plt.barh(y, x, color=(161/255, 48/255, 52/255))

# Label the x-axis and the y-axis
plt.xlabel('NOK', fontsize=14)
plt.ylabel('Linjeforening', fontsize=14)
plt.gcf().subplots_adjust(top=0.99, right=0.98)
plt.savefig("linjeforeningskonkurranse.png")
```
